```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['xtick.direction'] = 'out'  
matplotlib.rcParams['ytick.direction'] = 'out'  
import seaborn as sns
print('import done !')
```

```python
from sklearn.metrics import confusion_matrix
animal_labels = ["ant", "bird", "cat"]
y_true = ["cat", "ant", "cat", "cat", "ant", "bird"]
y_pred = ["ant", "ant", "cat", "cat", "ant", "cat"]
confusionmatrix = confusion_matrix(y_true, y_pred, labels=animal_labels)
```

```python
%matplotlib inline
fig, ax = plt.subplots(1)
ax = sns.heatmap(confusionmatrix, ax=ax, cmap=plt.cm.Greens, annot=True)
ax.set_xticklabels(animal_labels)
ax.set_yticklabels(animal_labels)
plt.title('Confusion Matrix')
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```
